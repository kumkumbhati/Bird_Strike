# Making the neccessary imports and preprocessing the data 

```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
pd.set_option('display.max_columns', None)
```

```python
#data cleaning and preprocessing
```

```python
data = pd.read_csv("Bird Strikes.csv")
data
```

```python
data = data.dropna(subset=['Is Aircraft Large?'])
data
```

```python
# Define a mapping dictionary for 'Y' and 'N'
yn_map = {'Y': True, 'N': False}

# Define a mapping dictionary for 'Yes' and 'No'
yes_no_map = {'Yes': True, 'No': False}

# Replace 'Y' and 'N' with boolean values in 'Pilot warned of birds or wildlife?'
data['Pilot warned of birds or wildlife?'] = data['Pilot warned of birds or wildlife?'].replace(yn_map)

# Replace 'Yes' and 'No' with boolean values in 'Is Aircraft Large?'
data['Is Aircraft Large?'] = data['Is Aircraft Large?'].replace(yes_no_map)

# Now the 'Pilot warned of birds or wildlife?' and 'Is Aircraft Large?' columns should contain boolean values
```

```python
data.head()
```

```python
data['FlightDate'] = pd.to_datetime(data['FlightDate'])
```

```python
data['Year'] = data['FlightDate'].dt.year
```

```python
data['Cost: Total $'] = data['Cost: Total $'].str.replace(',', '').astype(int)
```

# Visualization for the project 

```python
# number of bird strikes 
```

```python
yearly_strikes = data['Year'].value_counts().sort_index()
```

```python
yearly_strikes.plot(kind = 'bar')
plt.title('Number of bird strikes per year')
plt.xlabel('year')
plt.ylabel('Wildlife: Number Struck Actual')
plt.show()
```

```python
# cost per year
yearly_cost = data.groupby('Year')['Cost: Total $'].sum()

yearly_cost.plot(kind = 'line')
plt.title('Cost in $ per year')
plt.xlabel('year')
plt.ylabel('cost in $')
plt.show()
```

```python
# Top aircraft with maximum strikes
top_aircraft = data['Aircraft: Airline/Operator'].value_counts().head(10)

top_aircraft.plot(kind = 'pie')
plt.title('Top 10 Aircrafts')
plt.show()
```

```python
# Airports with maximum hits
airports = data['Airport: Name'].value_counts().head(30)

airports.plot(kind = 'bar')
plt.title('Maximum hits on Airports')
plt.xlabel('Airports')
plt.ylabel('Number of hits')
plt.show()
```

```python
#phase of flight 
phase = data['When: Phase of flight'].value_counts()
phase.plot(kind = 'pie', startangle=140)
plt.title('Bird Strikes by Phase of Flight')
plt.ylabel('')  
plt.xlabel('')  
plt.show()
```

```python
#cloud affecting strikes
cloud = data['Conditions: Sky'].value_counts()
cloud.plot(kind = 'bar')
plt.title('Cloud conditions')
plt.xlabel('Clouds')
plt.ylabel('Counts')
plt.show()
```

```python
#does sixe of the aircraft matters
size = data['Is Aircraft Large?'].value_counts()
size.plot(kind = 'pie',autopct = '%1.1f%%')
plt.title('Is Aircraft Large?')
plt.xlabel('')
plt.ylabel('')
plt.show()
```

```python
#was pilot aware of the situation 
aware = data['Pilot warned of birds or wildlife?'].value_counts()
aware.plot(kind = 'pie',autopct = '%1.1f%%')
plt.title('warning given by pilot')
plt.xlabel('')
plt.ylabel('')
plt.show()
```
